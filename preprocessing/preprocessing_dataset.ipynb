```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from collections import Counter
```

```python
nome_xls = '../wos_dataset/excel'
arquivos_xls = 22
nome_df = 'df_original_wos'

df_total = pd.concat([
    pd.read_excel(f"{nome_xls}{i}.xls")
    for i in range(1, arquivos_xls + 1)
], ignore_index=True)

df_total = df_total.dropna(subset='Abstract')
display(df_total)
```

```python
abstracts = list(df_total["Abstract"])
titulos = list(df_total["Article Title"])
dois = list(df_total["DOI"])
autores = list(df_total["Authors"])
anos = list(df_total["Publication Year"])
wos_id = wos_id = list(df_total["UT (Unique WOS ID)"])
artigo_id = []

for i in range(len(abstracts)):
    a = i + 1
    artigo_id.append(a)

resultados = {
    "title": titulos,
    "authors": autores,
    "abstract": abstracts,
    "DOI": dois,
    "year": anos,
    "article_id": artigo_id,
    "wos_id": wos_id
}

df = pd.DataFrame(resultados)

df.to_excel(f"../dataset/{nome_df}.xlsx", index=False)
print(f"Saved '{nome_df}.xlsx'.")
```
